# Linear Regression (Basic) - Experiment 01
Objective: Implement Linear Regression from scratch using synthetic data to understand the math behind it.

```python
import numpy as np
import matplotlib.pyplot as plt

# for reproducibility
np.random.seed(42)
```

```python
# generate synthetic linear data: y = 3x + 4 + noise
X = 2 * np.random.rand(100, 1)
y = 3 * X + 4 + np.random.rand(100, 1) # adding noise

# Visualize the data
plt.scatter(X, y, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic Data for Linear Regression')
plt.show()
```

```python
X = 2 * np.random.rand(100, 1)
noise = np.random.rand(100, 1)
y = 3 * X + 4 + noise

# show first 5 samples in a table-like format
for i in range(5):
    X_val = float(X[i, 0])
    ideal_y = float(3 * X_val + 4)
    noise_val = float(noise[i, 0])
    final_y = float(y[i, 0])
    print(f"Sample {i+1}: X={X_val:.3f}, Ideal_y={ideal_y:.3f}, Noise={noise_val:.3f}, Final_y={final_y:.3f}")
    
```

```python
# 1.14 : Plot Dataset
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel('Feature X')
plt.ylabel('Target y')
plt.title('Synthetic Linear Regression Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Hyperparameters
lr = 0.01 # learning rate
epochs = 1000 # Number of Iterations

# initialize parameters
w = 0.0
b = 0.0
n = len(X)

# Gradient Descent loop
for i in range(epochs):
    y_pred = w * X + b # predicted values
    dw = (-2/n) * np.sum(X * (y - y_pred)) # gradient wrt w
    db = (-2/n) * np.sum(y - y_pred)       # gradient wrt b
    w -= lr * dw                           # update w
    b -= lr * db                           # update b
    if i % 100 == 0:                       # print every 100 epochs
        mse = np.mean((y - y_pred)**2)
        print(f"Epoch {i}: MSE={mse:.4f}, w={w:.4f}, b={b:.4f}")
print(f"\nFinal parameters: w={w:.4f}, b={b:.4f}")
    
```
