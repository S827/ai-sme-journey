```python
# 1.Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 2. Generate same synthetic dataset
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1)
```

```python
# 3. Initialize and train model
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# 4. Extract parameters
print("Intercept (b):", lin_reg.intercept_[0])
print("Coefficient (w):", lin_reg.coef_[0][0])
```

```python
# 5. Predict
X_new = np.array([[0], [2]])
y_pred = lin_reg.predict(X_new)
```

```python
# 6. Plot
plt.scatter(X, y, alpha=0.5)
plt.plot(X_new, y_pred, color='red', linewidth=2, label="Best fit line (Sklearn)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# microstep 1.16
# 1. training data scatter plot
plt.scatter(X, y, color='blue', label='Training Data')

# 2. Regression line
X_new = np.array([[0], [2]]) # range for line
y_pred = lin_reg.predict(X_new)
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Prediction Line')

# 3. Labels and Legend
plt.xlabel('X')
plt.ylabel('y')
plt.title('linear Regression Fit')
plt.legend()
plt.show()
```
